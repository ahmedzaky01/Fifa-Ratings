```python
#We can group players by country and their overall, Potential and position so we can see who has the highest rating in America
def country(x):
    return fifa[fifa['Nationality'] == x][['Name','Overall','Potential','Position']]


# let's check the American Players 
country('United States')
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
sns.countplot(fifa['Preferred Foot'], palette = 'pink')
plt.title('Most Preferred Foot of the Players', fontsize = 20)
plt.show()
```

#### comparison of preferred foot over the different players

```python
plt.figure(figsize = (13, 8))
ax = sns.countplot(x = 'Height', data = fifa, palette = 'dark')
ax.set_title(label = 'Count of players on Basis of Height', fontsize = 20)
ax.set_xlabel(xlabel = 'Height in Foot per inch', fontsize = 16)
ax.set_ylabel(ylabel = 'Count', fontsize = 16)
plt.show()
```

#### as we can see 6' is the most common height between players

```python
#As of curiosity, I want to know the tallest five players
fifa[['Height', 'Name','Club']].sort_values('Height', ascending=False).head(5)
```

```python
#and the shortest five players
fifa[['Height', 'Name','Club']].sort_values('Height', ascending=True).head(5)
```

```python
plt.figure(figsize = (15, 7))
plt.style.use('_classic_test')

sns.countplot(x = 'Work Rate', data = fifa, palette = 'hls')
plt.title('Different work rates of the Players Participating in the FIFA 2019', fontsize = 20)
plt.xlabel('Work rates associated with the players', fontsize = 16)
plt.ylabel('count of Players', fontsize = 16)
plt.show()
```

```python
plt.style.use('dark_background')
fifa['Nationality'].value_counts().head(40).plot.bar(color = 'green', figsize = (20, 7))
plt.title('Different Nations Participating in FIFA 2019', fontsize = 30, fontweight = 20)
plt.xlabel('Name of The Country')
plt.ylabel('count')
plt.show()
```

```python
sns.set(style = "dark", palette = "pink", color_codes = True)
x = fifa.Age
plt.figure(figsize = (15,8))
ax = sns.distplot(x, bins = 58, kde = False, color = 'g')
ax.set_xlabel(xlabel = "Player\'s age", fontsize = 16)
ax.set_ylabel(ylabel = 'Number of players', fontsize = 16)
ax.set_title(label = 'Histogram of players age', fontsize = 20)
plt.show()
```
